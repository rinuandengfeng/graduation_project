```python
import pandas as pd
```

## 畅销榜图书数据合并

```python
# 合并畅销榜数据
bests2020_1 = pd.read_csv('../data/2020/bests_sellers_0_month.csv',header=None)
bests2021_1 = pd.read_csv('../data/2021/bests_sellers_0_month.csv',header=None)
bests2022_1 = pd.read_csv('../data/2022/bests_sellers_0_month.csv',header=None)
bests2023_1 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_0_month.csv',header=None)
bests2023_2 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_2_month.csv',header=None)
bests2023_3 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_3_month.csv',header=None)
bests2023_4 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_4_month.csv',header=None)
bests2023_5 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_5_month.csv',header=None)
bests2023_6 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_6_month.csv',header=None)
bests2023_7 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_7_month.csv',header=None)
bests2023_8 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_8_month.csv',header=None)
bests2023_9 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_9_month.csv',header=None)
bests2023_10 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_10_month.csv',header=None)
bests2023_11 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_11_month.csv',header=None)
bests2023_12 = pd.read_csv('../data/2023/bests_sellers/bests_sellers_12_month.csv',header=None)

bests_list = [bests2020_1,bests2021_1,bests2022_1,bests2023_1,bests2023_2,bests2023_3,bests2023_4,bests2023_5,bests2023_6,bests2023_7,bests2023_8,bests2023_9,bests2023_10,bests2023_11,bests2023_12]
```

### 删除第一行数据信息

```python
# 删除第一行数据信息
for bests in bests_list:
    print(bests + "删除第一行表头数据")
    bests.drop([0],inplace=True)
    print(bests + "删除第一行表头数据完成")
```

### 添加类型及时间两个列

```python
# 添加top_tag top_date两列
bests2020 = bests2020_1.assign(top_tag='bests',top_date='2020-01')
bests2021 = bests2021_1.assign(top_tag='bests',top_date='2021-01')
bests2022 = bests2022_1.assign(top_tag='bests',top_date='2022-01')
bests2023_1 = bests2023_1.assign(top_tag='bests',top_date='2023-01')
bests2023_2 = bests2023_2.assign(top_tag='bests',top_date='2023-02')
bests2023_3 = bests2023_3.assign(top_tag='bests',top_date='2023-03')
bests2023_4 = bests2023_4.assign(top_tag='bests',top_date='2023-04')
bests2023_5 = bests2023_5.assign(top_tag='bests',top_date='2023-05')
bests2023_6 = bests2023_6.assign(top_tag='bests',top_date='2023-06')
bests2023_7 = bests2023_7.assign(top_tag='bests',top_date='2023-07')
bests2023_8 = bests2023_8.assign(top_tag='bests',top_date='2023-08')
bests2023_9 = bests2023_9.assign(top_tag='bests',top_date='2023-09')
bests2023_10 = bests2023_10.assign(top_tag='bests',top_date='2023-10')
bests2023_11 = bests2023_11.assign(top_tag='bests',top_date='2023-11')
bests2023_12 = bests2023_12.assign(top_tag='bests',top_date='2023-12')
```

```python
bests2020.dtypes
```

```python
bests_total = pd.concat([bests2020, bests2021, bests2022, bests2023_1, bests2023_2, bests2023_3, bests2023_4, bests2023_5, bests2023_6, bests2023_7, bests2023_8, bests2023_9, bests2023_10, bests2023_11, bests2023_12])

bests_total.columns = ["top",'book_name','author','translater','book_image','publication_date','publication','current_price','pricing',
                       'discount','total_type','sub_type','book_size','paper','packaging','suit','isbn','editor_recommend','comment_number',
                       'comment_tag','top_tag','top_date']
```

```python
bests_total.head()
```

```python
bests_total.to_csv('../data/bests_total.csv',index=False)
```
